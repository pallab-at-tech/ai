```python
def is_magic_square(matrix):
    n = len(matrix)
    magic_sum = sum(matrix[0])

    # Check all rows
    for row in matrix:
        if sum(row) != magic_sum:
            return False

    # Check all columns
    for col in range(n):
        if sum(matrix[row][col] for row in range(n)) != magic_sum:
            return False

    # Check main diagonal
    if sum(matrix[i][i] for i in range(n)) != magic_sum:
        return False

    # Check secondary diagonal
    if sum(matrix[i][n - 1 - i] for i in range(n)) != magic_sum:
        return False

    return True


# Taking matrix input from user
n = int(input("Enter the size of the square matrix (n): "))
matrix = []

print("Enter the elements row by row:")
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

# Display the matrix
print("\nMatrix entered:")
for row in matrix:
    print(row)

# Check if it is a magic square
if is_magic_square(matrix):
    print("\nThe given matrix is a Magic Square.")
else:
    print("\nThe given matrix is NOT a Magic Square.")
```

```python
# Tic-Tac-Toe game in Python (2 players)

# Function to display the board
def print_board(board):
    print("\n")
    for row in board:
        print(" | ".join(row))
    print("-" * 9)
    print("\n")

# Function to check for a winner
def check_winner(board, player):
    # Check rows and columns
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or \
           all(board[j][i] == player for j in range(3)):
            return True

    # Check diagonals
    if all(board[i][i] == player for i in range(3)) or \
       all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

# Function to check if board is full
def is_full(board):
    return all(board[i][j] != ' ' for i in range(3) for j in range(3))

# Main game function
def tic_tac_toe():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    current_player = 'X'

    print("Welcome to Tic-Tac-Toe!")
    print_board(board)

    while True:
        print(f"Player {current_player}'s turn.")
        
        try:
            row = int(input("Enter row (1-3): ")) - 1
            col = int(input("Enter column (1-3): ")) - 1
        except ValueError:
            print("Invalid input! Please enter numbers between 1 and 3.")
            continue

        if 0 <= row < 3 and 0 <= col < 3:
            if board[row][col] == ' ':
                board[row][col] = current_player
                print_board(board)

                if check_winner(board, current_player):
                    print(f"ðŸŽ‰ Player {current_player} wins! ðŸŽ‰")
                    break
                elif is_full(board):
                    print("It's a draw!")
                    break
                else:
                    # Switch player
                    current_player = 'O' if current_player == 'X' else 'X'
            else:
                print("That spot is already taken! Try again.")
        else:
            print("Invalid position! Choose row and column between 1 and 3.")

# Run the game
tic_tac_toe()
```
