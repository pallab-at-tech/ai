```python
def dfs_implement(graph , start):
    visited = set()
    stack = [start]
    result = []

    while stack :
        node = stack.pop()
        if node not in visited :
            visited.add(node)
            result.append(node)

            for neighbors in  graph[node]:
                stack.append(neighbors)

    return result
```

```python
from collections import deque

def bfs_implement(graph , start):
    visited = set()
    result = []
    queue = deque([start])

    while queue:
        node = queue.popleft()

        if node not in visited:
            result.append(node)
            visited.add(node)

            for neighbors in graph[node]:
                queue.append(neighbors)

    return result
    
```

```python
graph = { 'A' : ['B','C'] , 'B' : ['D' , 'E'] , 'C' : ['F'] , 'E' : ['F'] , 'D' : [] , 'F' : [] }
print("DFS : ",dfs_implement(graph,'A'))
print("DFS : ",bfs_implement(graph,'A'))
```
